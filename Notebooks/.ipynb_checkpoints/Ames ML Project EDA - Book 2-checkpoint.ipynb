```python
# There is obvious relationship (as expected) in Overall Qual
# and SalePrice by Neighborhood - so I decided to divide the
# 'Neighborhood' var into groups using an interaction metric
# OverallQual * SalePrice
```

```python
df3['Qual_x_Sale'] = df3['OverallQual'] * df3['SalePrice']
```

```python
df3.head()
```

```python
df3 = df3.sort_values('Qual_x_Sale', ascending=False)
#change term to log - it's skewed
df3['Qual_x_Sale'] = np.log2(df3['Qual_x_Sale'])
```

```python
# plotting the interaction metric
df3['Qual_x_Sale'].describe()
```

```python
df3.plot(x='Neighborhood', y='Qual_x_Sale', kind='line', \
         title='Qual_x_Sale by Neighborhood')
plt.axhline(y=20.42, color='k', linestyle='--')
plt.axhline(y=20.17, color='k', linestyle='--')
plt.axhline(y=19.39, color='k', linestyle='--')
```

```python
# I will divide the neighborhoods by quartile using Qual_x_Sale
# Log transform - so labeled '1' (highest) to '4' (lowest)
# This will be a new column called 'NeighType'
```

```python
# category '1' (highest) are the first 7 neighborhoods
cat_1 = ['NoRidge', 'NridgHt', 'StoneBr', 'Timber', 'Somerst', 'Veenker', 'Blmgntn']

#category '2' (second highest) are next 6 neighborhoods
cat_2 = ['CollgCr', 'Crawfor', 'Gilbert', 'ClearCr', 'NWAmes', 'SawyerW']

#category '3' (second lowest) are next 6 neighborhoods
cat_3 = ['Mitchel', 'NPkVill', 'Blueste', 'NAmes', 'SWISU', 'OldTown']

#category '4' (lowest) are the final 6 neighborhoods
cat_4 = ['Sawyer', 'Edwards', 'BrkSide', 'BrDale', 'IDOTRR', 'MeadowV']
```

```python
# Now I need to add the column to the dataframe - in a way I can also add
# to the df_test later - I will start with a list of the lists &
# convert to a dictionary
cats = [cat_1, cat_2, cat_3, cat_4]
```

```python
cats
```

```python
cat_dict = {}
def cat_nums(cats):
    i = 4
    for n_list in cats:
        for n in n_list:
            cat_dict[n] = i
        i -= 1
cat_nums(cats)

# Now map the dictionary vals to a new column in the df
df_train['NeighType'] = df_train['Neighborhood'].map(cat_dict).astype(str)
```

```python
# need to make this a bit more readable - I'll remove noise
# will do a neg and pos threshold and plot 2 heatmaps overlaid
neg = df_train.corr()[df_train.corr() < -0.2]
pos = df_train.corr()[df_train.corr() > 0.4]
pos = pos[pos < 1.0]
fig, axs = plt.subplots(ncols=2, figsize=(20,10))
pos_heatmap = sns.heatmap(pos, center=0, linewidths=0.01, \
                          ax = axs[0])
pos_heatmap.set(facecolor="#D6F2EB")
pos_heatmap.set_title('Positive +0.4 Correlations', fontsize=20)
neg_heatmap = sns.heatmap(neg, center=0, linewidths=0.01, \
                         ax = axs[1])
neg_heatmap.set(facecolor="#F2D6DD")
neg_heatmap.set_title('Negative -0.2 Correlations', fontsize=20)
```

```python
corr = df_train.corr()
mask = np.triu(np.ones_like(corr, dtype=bool))
f, ax = plt.subplots(figsize=(11,9))
cmap = sns.diverging_palette(230, 20, as_cmap=True)
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=0.3, center=0, \
           square=True, linewidths=0.5, cbar_kws={'shrink': 0.5})
ax.set_title('Continuous Features - Correlations')
```

```python
neg = df_train.corr()[df_train.corr() < -0.2]
pos = df_train.corr()[df_train.corr() > 0.4]
pos = pos[pos < 1.0]

fig, axs = plt.subplots(ncols=2, figsize=(20,10))
cmap = sns.diverging_palette(230, 20, as_cmap=True)

pos_mask = np.triu(np.ones_like(pos, dtype=bool))
pos_heatmap = sns.heatmap(pos, mask=pos_mask, cmap=cmap, vmax=0.3, \
                          center=0, square=True, linewidths=0.05, \
                          cbar_kws={'shrink' :0.5}, ax = axs[0])
pos_heatmap.set(facecolor="#EAFAF8")
pos_heatmap.set_title('Positive +0.4 Correlations', fontsize=14)

neg_mask = np.triu(np.ones_like(neg, dtype=bool))
neg_heatmap = sns.heatmap(neg, mask=neg_mask, cmap=cmap, vmax=0.3, \
                          center=0, square=True, linewidths=0.05, \
                          cbar_kws={'shrink' :0.5}, ax = axs[1])
neg_heatmap.set(facecolor="#FDE5DF")
neg_heatmap.set_title('Negative -0.2 Correlations', fontsize=14)
```
