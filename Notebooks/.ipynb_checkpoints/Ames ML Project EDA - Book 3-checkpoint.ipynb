```python
#This notebook has steps for data cleanup, correcting skewness,
#dummies and categorical vars, and some feature engineering
import numpy as np
import pandas as pd
pd.set_option('display.max_columns', 100)

import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns
sns.set_style('darkgrid')

path_to_train = '~/Desktop/Ames_ML_Project/Data/train_checkpoint_df_01_16.csv'
path_to_test = '~/Desktop/Ames_ML_Project/Data/test_checkpoint_df.csv'
```

```python
#load test data and align format w/ train dataframe (book 2)
df_test = pd.read_csv(path_to_test, index_col=False)
df_train = pd.read_csv(path_to_train, index_col=False)
```

```python
df_test.drop(['Unnamed: 0'], axis=1, inplace=True)
df_test.drop(['Id'], axis=1, inplace=True)
df_train.drop(['Unnamed: 0'], axis=1, inplace=True)
```

```python
import matplotlib

params = {'axes.titlesize' : '18',
         'xtick.labelsize' : '14',
         'ytick.labelsize' : '14'}
matplotlib.rcParams.update(params)
df_train.hist(figsize=(14,14))
plt.tight_layout()
plt.show()
```

```python
# Drop vars with very little info or duplicate info:
to_drop = ['3SsnPorch', 'EnclosedPorch', 'GarageYrBlt', 'LotArea', \
          'KitchenAbvGr', 'LowQualFinSF', 'PoolArea', 'ScreenPorch']
df_train.drop(to_drop, axis=1, inplace=True)
```

```python
# Find correlation of GarageArea and GarageCars
df_train['GarageArea'].corr(df_train.GarageCars)
```

```python
# Let's drop the latter
df_train.drop('GarageCars', axis=1, inplace=True)
```

```python
# Seems like there are too many bathroom vars - combine them
BsmtFull = df_train['BsmtFullBath']
BsmtHalf = df_train['BsmtHalfBath']
FullBath, HalfBath = df_train['FullBath'], df_train['HalfBath']
df_train['Baths'] = (FullBath + BsmtFull) + 0.5*(HalfBath + BsmtHalf)
```

```python
# Now drop the other columns
df_train.drop(['BsmtFullBath', 'BsmtHalfBath', 'FullBath', 'HalfBath'], \
             axis = 1, inplace = True)
```

```python
# Create a second var for '2ndFloor' Y/N 
def f(row):
    if row['2ndFlrSF'] > 0:
        val = 'Y'
    else:
        val = 'N'
    return val
df_train['2Story'] = df_train.apply(f, axis=1)
```

```python
# Drop 2ndFlr SF - so we can deal with SF next
df_train.drop('2ndFlrSF', axis=1, inplace=True)
```

```python
corr = df_train.corr()
print(corr['GrLivArea'].sort_values(ascending=False)[1:10])
```

```python
print(corr['TotRmsAbvGrd'].sort_values(ascending=False)[1:10])
```

```python
# Because I'm keeping GrLivArea and I have '2Story' var I will drop
# 1stFlrSF
df_train.drop('1stFlrSF', axis=1, inplace=True)
```

```python
basement_cols = df_train.columns[df_train.columns.str.contains('Bsmt')]
```

```python
df_train[basement_cols].describe()
```

```python
# Let's create 'BsmtFinSF' and keep 'TotalBsmtSF'
BsmtFin1, BsmtFin2 = df_train['BsmtFinSF1'], df_train['BsmtFinSF2']
BsmtUnfSF, TotalBsmtSF = df_train['BsmtUnfSF'], df_train['TotalBsmtSF']
```

```python
df_train['BsmtFinSF'] = TotalBsmtSF - BsmtUnfSF
```

```python
df_train.drop(['BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF'], axis=1, inplace=True)
```

```python
# All the size vars are skewed because of relatively few very large houses
# transform the following using box/cox or log:
skewed_vars = ['GarageArea', 'BedroomAbvGr', 'TotRmsAbvGrd', 'GrLivArea', \
                   'LotFrontage', 'TotalBsmtSF', 'BsmtFinSF']
```

```python
from scipy.special import boxcox
l = 0.15
for feature in skewed_vars:
    df_train[feature] = boxcox(df_train[feature], l)
```

```python
# also unskew 'Baths'
df_train['Baths'] = boxcox(df_train['Baths'], l)
```

```python
# Checking to see if the transformation reduced skew
import matplotlib

params = {'axes.titlesize' : '32',
         'xtick.labelsize' : '24',
         'ytick.labelsize' : '24'}
matplotlib.rcParams.update(params)
df_train.hist(figsize=(50,30))
plt.show()
```

```python
# Create a dictionary to map values above to values for dummification (condolidated etc)
cat_dict = {}
to_Oth = ['RM', 'FV', 'RH', 'Bnk', 'Low', 'HLS', 'FR2', 'Corner', 'CulDSac', 'FR3', \
           'Mod', 'Sev', '2fmCon', 'Duplex', 'TwnhsE', 'Twnhs', 'Hip', 'Gambrel', 'Mansard', \
           'Flat', 'Shed', 'BrkTil', 'Wood', 'Slab', 'Stone', 'Detchd', 'BuiltIn', 'CarPort', \
           'Basment']
to_Yes = ['Feedr', 'PosN', 'Artery', 'RRAe', 'RRAn', 'PosA', 'RRNe', 'RRNn', 'Y', 'SBrkr', \
          'MnPrv', 'GdWo', 'GdPrv', 'MnWw']
to_None = ['0', 'N', 'Unf', 'FuseF', 'FuseA', 'FuseP', 'Mix', 'P']
to_Avg = ['TA', 'Po', 'Fa']
to_Gd = ['Gd', 'Ex']
to_Brick = ['BrkFace', 'BrkCmn']
to_Attchd = ['Attchd', '2Types']
to_Self = ['RL', 'Reg', 'Lvl', 'Inside', 'Gtl', '1Fam', 'Gable', 'PConc', 'CBlock', \
          'GLQ', 'ALQ', 'Rec', 'BLQ', 'LwQ']
```

```python
# Now build the dictionary and map to df_train
cat_dict.update(dict.fromkeys(to_Other, 'Oth'))
cat_dict.update(dict.fromkeys(to_Yes, 'Yes'))
cat_dict.update(dict.fromkeys(to_None, 'None'))
cat_dict.update(dict.fromkeys(to_Avg, 'Avg'))
cat_dict.update(dict.fromkeys(to_Gd, 'Gd'))
cat_dict.update(dict.fromkeys(to_Brick, 'Brick'))
cat_dict.update(dict.fromkeys(to_Attchd, 'Attchd'))
```

```python
map_cols = ['MSZoning', 'LotShape', 'LandContour', 'LotConfig', 'LandSlope', 'BldgType', \
           'RoofStyle', 'MasVnrType', 'Foundation', 'BsmtQual', 'BsmtFinType1', 'HeatingQC', \
           'CentralAir', 'Electrical', 'KitchenQual', 'GarageType', 'PavedDrive', 'Fence']
```

```python
for col in map_cols:
    df_train[col] = df_train[col].map(cat_dict).fillna(df_train[col])
```

```python
# Condition2 has no value - Condition1 is either 'Norm'='No' or Other='Yes'
# Creating new var called 'RoadRail' which is Yes/No for close to one of these (bad) things
df_train.drop('Condition2', axis=1, inplace=True)
cond_dict = {}
to_None = ['Norm']
to_Yes = ['Feedr', 'Artery', 'RRAn', 'PosN', 'RRAe', 'PosA', 'RRNn', 'RRNe']
cond_dict.update(dict.fromkeys(to_None, 'None'))
cond_dict.update(dict.fromkeys(to_Yes, 'Yes'))
df_train['RoadRail'] = df_train['Condition1'].map(cond_dict).fillna(df_train['Condition1'])
```
